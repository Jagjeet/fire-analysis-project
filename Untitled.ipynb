```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import sqlite3
import sqlalchemy as sqla
import scipy.stats as st
from scipy.stats import linregress
from api_keys import air_key
```

```python

con = sqlite3.connect("./Resources/FPA_FOD_20170508.sqlite")
fire_data = pd.read_sql_query("SELECT OBJECTID, FOD_ID, FPA_ID, FIRE_YEAR, FIPS_NAME, FIRE_NAME, FIRE_SIZE, DISCOVERY_DATE, FIRE_CODE, STAT_CAUSE_DESCR, CONT_DATE, STATE, LATITUDE, LONGITUDE, DISCOVERY_TIME, CONT_TIME from Fires", con)

print(fire_data)

con.close()
```

```python
fire_data = fire_data.dropna(inplace = False)
fire_data['Start Date'] = (pd.to_datetime((fire_data.DISCOVERY_DATE // 1000).astype(int)) + 
                 pd.to_timedelta(fire_data.DISCOVERY_DATE % 1000, unit='D'))
fire_data['End Date'] = (pd.to_datetime((fire_data.CONT_DATE // 1000).astype(int)) + 
                 pd.to_timedelta(fire_data.CONT_DATE % 1000, unit='D'))
fire_data["Start Date"] = pd.DatetimeIndex(fire_data['Start Date']).date
fire_data["End Date"] = pd.DatetimeIndex(fire_data['End Date']).date

fire_data
```

```python
# time_strs = fire_data["DISCOVERY_TIME"].astype(str).str.zfill(4)
# fire_data["DISCOVERY_TIME"] = pd.to_datetime(time_strs, format="%H%M")

# time_strs2 = fire_data["CONT_TIME"].astype(str).str.zfill(4)
# fire_data["CONT_TIME"] = pd.to_datetime(time_strs2, format="%H%M")

# fire_data["DISCOVERY_TIME"] = pd.DatetimeIndex(fire_data['DISCOVERY_TIME']).time
# fire_data["CONT_TIME"] = pd.DatetimeIndex(fire_data['CONT_TIME']).time
fire_data.head()
```

```python
fire_data["Start Date"] = fire_data.apply(lambda r : pd.datetime.combine(r['Start Date'],r['DISCOVERY_TIME']),1)
fire_data["End Date"] = fire_data.apply(lambda r : pd.datetime.combine(r['End Date'],r['CONT_TIME']),1)
fire_data.head()
```

```python
cause = fire_data.groupby(["STAT_CAUSE_DESCR"])
cause = cause.count()
cause = cause[["FIRE_NAME"]]
cause = cause.rename(columns = {"FIRE_NAME": "Number of Fires"})
cause = cause.sort_values(by=['Number of Fires'])
cause
```

```python
pie_cause = cause.plot(kind = "pie", y="Number of Fires", labeldistance = None, figsize=(20,10), fontsize=9, autopct='%1.0f%%', pctdistance=1.1, title= "Cause of Fires from 1992 to 2015")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig("output/Figure1")

plt.show()
```

```python
fire_data["Days_Lasted"] = fire_data["End Date"] - fire_data["Start Date"]

fire_data['Total Minutes'] = fire_data['Days_Lasted'].dt.total_seconds().div(60).astype(int)
fire_data
```

```python
fire_size_date = fire_data.loc[fire_data["Total Minutes"]>0, :]
fire_num = fire_size_date[["FIRE_YEAR", "Total Minutes"]]
yearly_fire_num = fire_num.groupby(["FIRE_YEAR"])
sum_fire_time = yearly_fire_num.sum() 
sum_fire_time
```

```python
ax4 = sum_fire_time.plot.line()
plt.xlabel('Year')
plt.ylabel('Total Minutes All Fires Burned')
ax4.get_legend().remove()
plt.savefig("output/Figure9")
```

```python
fire_size_date = fire_data.loc[fire_data["Total Minutes"]>0, :]

minutes_last = fire_size_date["Total Minutes"]
fire_size = fire_size_date["FIRE_SIZE"]
plt.scatter(x = minutes_last, y = fire_size, c = "DarkBlue")


correlation = st.pearsonr(minutes_last,fire_size)
print(f"The correlation between size and time is {round(correlation[0],2)}")

(slope, intercept, rvalue, pvalue, stderr) = linregress(minutes_last,fire_size)
regress_values = minutes_last * slope + intercept
line = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.plot(minutes_last,regress_values,"r-")

plt.xlabel('Minutes Lasted')
plt.ylabel('Acres Burned')
plt.savefig("output/Figure10")
```
